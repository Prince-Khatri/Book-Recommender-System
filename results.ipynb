```python
import matplotlib.pyplot as plt
%matplotlib inline
# K values
k_values = [5, 10, 15, 20]

# RMSE values
user_cf_rmse = [0.8607, 0.8846, 0.8601, 0.8801]
item_cf_rmse = [0.8564, 0.8628, 0.8530, 0.8576]
hybrid_rmse = [0.8078, 0.7971, 0.7923, 0.8002]

# Create figure
plt.figure(figsize=(10, 6))

# Plot lines
plt.plot(k_values, user_cf_rmse, marker='o', linewidth=2, label='User-Based CF')
plt.plot(k_values, item_cf_rmse, marker='s', linewidth=2, label='Item-Based CF')
plt.plot(k_values, hybrid_rmse, marker='^', linewidth=2, label='Hybrid Ensemble')

# Labels and title
plt.xlabel('K Value', fontsize=12)
plt.ylabel('RMSE', fontsize=12)
plt.title('RMSE vs K for Recommendation Models (20K Dataset)', fontsize=14)

# Show all K values on x-axis
plt.xticks(k_values)

# Grid and legend
plt.grid(True)
plt.legend()

# Show graph
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Models and RMSE values
models = ['User-Based CF', 'Item-Based CF', 'Hybrid Ensemble']
rmse_values = [1.0628, 0.6959, 0.7935]

# Create line chart
plt.figure(figsize=(9, 5))

plt.plot(
    models,
    rmse_values,
    marker='o',
    linewidth=2
)

# Add RMSE labels on points
for i, value in enumerate(rmse_values):
    plt.text(i, value + 0.01, f'{value:.4f}', ha='center')

# Labels and title
plt.xlabel('Models')
plt.ylabel('RMSE')
plt.title('Model vs RMSE on Full Dataset (3M Ratings)')

# Grid
plt.grid(True)

# Show plot
plt.tight_layout()
plt.show()
```
